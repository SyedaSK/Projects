### Aim<a id='aim'></a>

This report aims to evaluate the performance of machine learning algorithms in predicting the presence or absence of heart disease in individuals based on their clinical and demographic features. Utilising Python and the Scikit-Learn module, we will analyse a dataset containing patients' medical histories, symptoms, and diagnostic test results to identify the most effective machine learning models for heart disease diagnosis. ### Literature Review<a id='literature'></a>

Identifying heart disease remains a critical focus in medical research and healthcare, as emphasized by Shorewala (2021). This urgency stems from the largely unknown causes of heart diseases, which present significant risks of mortality, severe morbidity, and disability, as highlighted by Subramani et al. (2023). Consequently, patients and governments worldwide have faced substantial socioeconomic burdens associated with addressing this pressing health issue.

Diagnosing heart disease is a complex and challenging task in the medical field, requiring a high degree of accuracy and efficiency. Early detection of heart disease can significantly reduce the risk of death. Given the increasing prevalence of cardiac issues, predicting heart disease has become one of the most challenging medical endeavors in recent years (MohiUddin, 2023). Consequently, researchers and the medical industry are actively exploring diverse approaches to identify markers that can reliably predict the onset of heart diseases.

The advent of digital world has led to the widespread use of electronic medical records in the healthcare industry, resulting in the generation of vast amounts of medical data for research purposes (Baghdadi et al., 2023). These datasets exhibit characteristics of heterogeneity and multi-dimensionality, offering valuable insights into the identification of markers associated with heart diseases (Chen et al., 2022). Thus, the integration of machine learning techniques shows great promise in advancing clinical practices for managing cardiovascular diseases.

Machine learning and big data analytics have the potential to enhance the accuracy of predictions made by doctors and medical professionals for their patients. Recent research has demonstrated promising outcomes, indicating that with sufficient clinical data, these technologies can predict the presence of heart diseases in patients (Ogunpola et al., 2024).

Porto et al. (2021) utilized the statlog dataset, which is also part of the dataset used in this project, to develop a cardiovascular disease prediction system. They focused on reducing the number of features, making the predictions easier to interpret. Additionally, they emphasized having an explainable model while keeping the predictions accurate. Ogundepo & Yahya (2023) conducted exploratory analysis and compared 10 common classification models for better prediction. They used Cleaveland dataset to build the model and statlog dataset to validate the results to demonstrate the robustness and reliability of the results. Their results concluded that Support Vector Machine (SVM) provided the best performance in terms of accuracy as well as balance between sensitivity and specificity.

Karadeniz, Tokdemir & Maraş (2021) used ensemble method combining multiple models to improve prediction accuracy. Their results showed 88.7% accuracy for Spectf dataset and 88.8% accuracy on the Statlog dataset. However, the model performed better on specific dataset and exhibited random state dependency. Rasool et al. (2020) used logistic regression (LR), Artificial Neural Network (ANN) and Support Vector Classifier (SVC) to analyse the statlog dataset with 270 unique instances. In this study SVC outperformed LN and ANN, having a prediction accuracy of 92%.

Anuradha & David (2021) used the Cleaveland, Statlog heart and SA heart to investigate the impact of feature selection on computational efficiency and accuracy of prediction. They used gradient boosting algorithms, CatBoost and XGBoost, to gain insights into optimal feature ranking that leads highest accuracy in  heard disease prediction.  Hera, Amjad & Saba (2022) proposed the used of multi-tier ensemble model with Random forest for feature selection. Their proposed model outperformed the results in other studies achieving 93.76% accuracy. They used Wilcoxon signed-rank test to confirm the significance of the results.

Sajeev et al. (2019) used both deep learning and machine learning methodologies to predict the presence of heart disease. They compared deep learning to linear regression, linear discriminant analysis, support vector machine (SVF) and random forest. Their results show deep learning model performs better than both linear and non-linear models used.

Researchers have employed various machine learning techniques to improve the precision of heart disease predictions (Jain et al., 2023). These methods include Logistic Regression, Artificial Neural Networks (ANN), Support Vector Classifier (SVC), CatBoost, XGBoost, Convolutional Neural Networks, and Random Forest. By investigating different models and assessing their performance, researchers have gleaned valuable insights aimed at developing a robust prediction model for accurately predicting heart disease (Roy et al., 2023).

This literature review concludes that applying machine learning to diagnose heart diseases is both practical and feasible, considering the abundance of health data currently accessible. Machine learning techniques can be utilized to construct heart disease risk prediction models, enabling doctors and healthcare professionals to assess patients' risk of heart disease with a high degree of accuracy, thereby enhancing clinical decision-making processes. Hence, it is crucial to evaluate and determine the most effective machine learning model for predicting heart diseases.


### Dataset Source

The heart disease dataset used in this study was sourced from IEEE DataPort (Siddhartha, 2020). Siddhartha curated this dataset by combining five independent heart disease datasets—Cleveland, Hungarian, Switzerland, Long Beach VA, and Statlog(Heart)—to create a larger dataset for analysis (Janosi et al., n.d.; Newman et al., n.d.).



### Dataset Details <a id='dataset-details'></a>

This dataset contains clinical and demographic information about patients, along with assessments indicating the presence or absence of heart disease.

The dataset includes various clinical data points such as chest pain type, resting blood pressure, serum cholesterol, fasting blood sugar, resting electrocardiogram results, maximum heart rate achieved, exercise-induced angina, old peak depression, and the slope of the peak exercise ST segment (Siddhartha, 2020). Additionally, it contains information on the age and sex of the patients. Each record in the dataset indicates whether the patient has heart disease. These features are considered comprehensive for applying machine learning algorithms to diagnose heart disease (Liu, 2017). Therefore, in this study, we will attempt predictive modeling of heart disease as a classification problem.


The dataset initially contains 1190 records with 11 features. However, for this study, records containing missing values and incorrect values are removed, specifically those with cholesterol values of 0, resting heart rate of 0, and old peak depression values that are negative. After this cleaning process, the dataset contains 745 records with 11 descriptive features and a target feature.

```python
# Masking rows where the value of resting_bp_s or cholesterol is 0
# And checking the sum
zero_mask = (data['resting_bp_s']==0) | (data['cholesterol']==0) | (data['oldpeak'] <0)
zero_mask.sum()
```
